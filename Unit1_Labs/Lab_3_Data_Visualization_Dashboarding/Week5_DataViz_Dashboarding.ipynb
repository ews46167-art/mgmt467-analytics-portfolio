### Q1 ‚Äî Exploratory vs. Explanatory
**Prompt:** *"Explain the difference between exploratory and explanatory visualization."*  
**Hints:** Consider audience, purpose, level of polish, and how much iteration you expect.

üìù **Your response:**

Exploratory Visualization:
Audience: Typically the analyst or data scientist.
Purpose: To understand the data, identify patterns, test hypotheses, and find insights. It's about discovery.
Level of Polish: Often rough and unpolished. Quick iterations are common.
Iteration: High. You might create many different plots quickly to see what the data reveals.
Explanatory Visualization:
Audience: A specific audience you want to communicate findings to (e.g., executives, clients, the public).
Purpose: To tell a specific story or communicate a clear message based on findings. It's about explaining.
Level of Polish: High. Visuals are carefully designed, cleaned up, and annotated to be easily understood by the target audience.
Iteration: Lower, focused on refining the chosen visuals to be as clear and persuasive as possible.

### Q2 ‚Äî Chart Types & When to Use Them
**Prompt:** *"Describe 5 common chart types (e.g., bar, line, scatter, histogram, map) and the type of data analysis question each is best suited to answer."*  

For each chart, list: **Question archetype**, **Data shape**, **Pitfalls**, **Alternatives**.

üìù **Your response (template):**

**Bar Chart:**
- Question archetype: How do values compare across different categories? (Comparison)
- Data shape: One categorical variable and one quantitative variable.
- Pitfalls: Too many categories can make it cluttered; misleading y-axis scale (not starting at 0).
- Alternatives: Column chart, stacked bar chart, Marimekko chart.

**Line Chart:**
- Question archetype: How does a value change over time or a continuous sequence? (Trend)
- Data shape: One temporal or ordered variable and one quantitative variable.
- Pitfalls: Too many lines can be hard to follow; misleading scale.
- Alternatives: Area chart, sparkline.

**Scatter Plot:**
- Question archetype: Is there a relationship or correlation between two quantitative variables? (Relationship)
- Data shape: Two quantitative variables.
- Pitfalls: Overplotting with too many data points; misinterpreting correlation as causation.
- Alternatives: Bubble chart (adds a third quantitative variable), heat map (for aggregated relationships).

**Histogram:**
- Question archetype: How is a single quantitative variable distributed? (Distribution)
- Data shape: One quantitative variable.
- Pitfalls: Bin size can significantly alter the appearance of the distribution; not suitable for comparing distributions of different categories easily.
- Alternatives: Density plot, box plot.

**Map:**
- Question archetype: How do values vary geographically? (Spatial)
- Data shape: Geographical data (latitude/longitude, regions, countries) and one or more quantitative or categorical variables.
- Pitfalls: Can be misleading if not normalized by population or area; geographical areas can distort perception of values (larger areas appearing more important).
- Alternatives: Choropleth map, symbol map, dot distribution map.

### Q3 ‚Äî Airline Executive Dashboard KPIs
**Prompt:** *"Your goal is to create a dashboard for an airline executive. Brainstorm 5 Key Performance Indicators (KPIs) that should be on this dashboard. For each KPI, suggest the best chart type to visualize it."*

Suggested KPI ideas to consider (customize them): On-Time Rate, Average Delay Minutes, Cancellation Rate, Load Factor, Yield (Revenue per Passenger Mile), Completion Factor, Net Promoter Score (if available), Revenue, Cost per Available Seat Mile (CASM).

üìù **Your response (template):**

**KPI: On-Time Rate**
- Why it matters: Directly impacts customer satisfaction and operational efficiency. Higher on-time rates often lead to happier passengers and lower costs associated with delays (rebooking, crew issues, etc.).
- Best chart: Bar chart (comparing carriers or routes), Line chart (trend over time).
Drill-down fields: Carrier, Origin-Dest Route, Time of Day, Day of Week, Month.
KPI: Average Arrival Delay Minutes
Why it matters: Quantifies the impact of delays. Lower average delay indicates smoother operations and better adherence to schedule.
Best chart: Line chart (trend over time, perhaps with a rolling average), Histogram (distribution of delays), Bar chart (average delay by carrier or route).
- Drill-down fields: Carrier, Origin-Dest Route, Airport (Origin or Destination), Time of Day, Month.

**KPI: Load Factor**
- Why it matters: Measures how full the planes are. Higher load factors generally mean more revenue per flight, as fixed costs are spread across more passengers.
- Best chart: Line chart (trend over time), Bar chart (by carrier or route), Scatter plot (Load Factor vs. Revenue or Cost).
- Drill-down fields: Carrier, Origin-Dest Route, Aircraft Type, Month, Day of Week.

**KPI: Revenue**
- Why it matters: Fundamental measure of financial performance. Executives need to track revenue streams to understand profitability and identify areas for growth.
- Best chart: Line chart (trend over time), Bar chart (by carrier, route, or region), Stacked bar chart (revenue by fare class).
- Drill-down fields: Carrier, Origin-Dest Route, Fare Class, Month, Sales Channel.

**KPI: Cost per Available Seat Mile (CASM)**
- Why it matters: A key metric for operational efficiency and cost management. Lower CASM indicates the airline is operating more cost-effectively.
- Best chart: Line chart (trend over time), Bar chart (by carrier or aircraft type), Scatter plot (Cost vs. Revenue, colored by CASM).
- Drill-down fields: Carrier, Aircraft Type, Origin-Dest Route, Fuel Costs, Maintenance Costs, Month.

### Create a Synthetic Airline Dataset (if you don't have one)
Use this dataset to prototype visualizations and dashboards. It contains realistic columns for the KPIs you defined. You can always swap in your BigQuery data later.


````python
# Generate a synthetic airline dataset
import pandas as pd
import numpy as np
from datetime import datetime, timedelta

np.random.seed(467)

dates = pd.date_range(start='2024-01-01', end='2024-06-30', freq='D')
carriers = ['AA', 'DL', 'UA', 'WN', 'AS']
origins = ['ATL','DFW','DEN','ORD','LAX','SEA','SFO','PHX','CLT','MCO']
dests =   ['JFK','LGA','BOS','SFO','LAX','SEA','MIA','IAH','IAD','PHL']

rows = []
for d in dates:
    for carrier in carriers:
        # number of flights per day per carrier
        n = np.random.randint(80, 140)
        for _ in range(n):
            o = np.random.choice(origins)
            de = np.random.choice(dests)
            if o == de:
                continue
            distance = np.random.randint(300, 2800)
            seats = np.random.choice([90,120,150,180,210])
            load_factor = np.clip(np.random.normal(0.82, 0.07), 0.55, 0.98)
            pax = int(seats * load_factor)
            dep_delay = int(np.random.normal(8, 15))
            arr_delay = dep_delay + int(np.random.normal(0, 12))
            dep_delay = max(dep_delay, -5)
            arr_delay = max(arr_delay, -10)
            cancelled = np.random.rand() < 0.025
            diverted = np.random.rand() < 0.005
            revenue = pax * np.random.uniform(120, 420)
            casm = np.random.uniform(0.08, 0.16)  # cost per available seat mile
            cost = casm * seats * distance
            fuel_burn = distance * np.random.uniform(3.0, 4.5)  # gallons per seat-mile proxy
            nps = np.random.normal(35, 15)  # if you want a customer metric proxy
            rows.append({
                'date': d,
                'carrier': carrier,
                'origin': o,
                'dest': de,
                'distance_miles': distance,
                'seats': seats,
                'passengers': pax,
                'dep_delay_min': dep_delay,
                'arr_delay_min': arr_delay,
                'cancelled': int(cancelled),
                'diverted': int(diverted),
                'revenue_usd': round(revenue, 2),
                'cost_usd': round(cost, 2),
                'fuel_burn_gal': round(fuel_burn, 1),
                'nps': round(nps, 1)
            })

df_flights = pd.DataFrame(rows)

# Derived KPIs
df_flights['on_time'] = (df_flights['arr_delay_min'] <= 15).astype(int)
df_flights['load_factor'] = (df_flights['passengers'] / df_flights['seats']).round(3)
df_flights['yield_usd_per_pax_mile'] = (df_flights['revenue_usd'] / (df_flights['passengers'] * df_flights['distance_miles']).clip(lower=1)).round(4)
df_flights['asm'] = df_flights['seats'] * df_flights['distance_miles']  # available seat miles
df_flights['casm'] = (df_flights['cost_usd'] / df_flights['asm']).round(4)
df_flights['month'] = df_flights['date'].dt.to_period('M').astype(str)

df_flights.head()
````

#### Prompt E ‚Äî Cost vs. Revenue with CASM Coloring (Scatter)
```
Given `df_flights` with `cost_usd`, `revenue_usd`, and `casm`, produce a **scatter plot** of cost vs. revenue with points colored by `casm` (continuous colormap) and add a diagonal y=x reference line. Include labels and legend.
```


````python
# Prompt A: On-Time Rate by Carrier (Bar Chart)
on_time_rate = df_flights.groupby('carrier')['on_time'].mean().sort_values(ascending=False) * 100

plt.figure(figsize=(10, 6))
ax = sns.barplot(x=on_time_rate.index, y=on_time_rate.values, palette='viridis')

plt.title('On-Time Rate by Carrier', fontsize=16)
plt.xlabel('Carrier', fontsize=12)
plt.ylabel('On-Time Rate (%)', fontsize=12)
plt.ylim(0, 100) # Ensure the y-axis starts at 0 and goes up to 100

# Add percentage labels on bars
for p in ax.patches:
    ax.annotate(f'{p.get_height():.1f}%', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', xytext=(0, 5), textcoords='offset points')

plt.tight_layout()
plt.show()
````

````python
# Prompt B: Average Arrival Delay over Time (Line Chart)
daily_avg_delay = df_flights.groupby('date')['arr_delay_min'].mean().reset_index()
daily_avg_delay['rolling_avg_delay'] = daily_avg_delay['arr_delay_min'].rolling(window=7).mean()

plt.figure(figsize=(12, 6))
sns.lineplot(data=daily_avg_delay, x='date', y='arr_delay_min', label='Daily Average Delay', color='skyblue')
sns.lineplot(data=daily_avg_delay, x='date', y='rolling_avg_delay', label='7-Day Rolling Average', color='salmon')

plt.title('Average Arrival Delay Over Time', fontsize=16)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Average Arrival Delay (minutes)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()
plt.tight_layout()
plt.show()
````

````python
# Prompt C: Delay Distribution (Histogram)
plt.figure(figsize=(10, 6))
sns.histplot(df_flights['arr_delay_min'], bins=50, kde=True)

plt.title('Distribution of Arrival Delays', fontsize=16)
plt.xlabel('Arrival Delay (minutes)', fontsize=12)
plt.ylabel('Frequency', fontsize=12)

# Add vertical line at 0 delay
plt.axvline(0, color='red', linestyle='--', label='On-Time Arrival')

# Calculate percentage of flights within [-5, 15] minutes
on_time_flights = df_flights[(df_flights['arr_delay_min'] >= -5) & (df_flights['arr_delay_min'] <= 15)]
percentage_on_time = (len(on_time_flights) / len(df_flights)) * 100

# Add annotation
plt.text(0.95, 0.95, f'Flights within [-5, 15] min: {percentage_on_time:.1f}%',
         horizontalalignment='right', verticalalignment='top', transform=plt.gca().transAxes,
         fontsize=10, bbox=dict(boxstyle='round,pad=0.5', fc='white', alpha=0.5))

plt.legend()
plt.tight_layout()
plt.show()
````

````python
# Prompt D: Load Factor by Route (Top 15) (Horizontal Bar)
df_flights['route'] = df_flights['origin'] + '-' + df_flights['dest']
route_load_factor = df_flights.groupby('route')['load_factor'].mean().sort_values(ascending=False).head(15) * 100

plt.figure(figsize=(10, 8))
sns.barplot(x=route_load_factor.values, y=route_load_factor.index, palette='viridis')

plt.title('Top 15 Routes by Average Load Factor', fontsize=16)
plt.xlabel('Average Load Factor (%)', fontsize=12)
plt.ylabel('Route', fontsize=12)
plt.xlim(0, 100) # Ensure the x-axis is a percentage

plt.tight_layout()
plt.show()
````

````python
# Prompt E: Cost vs. Revenue with CASM Coloring (Scatter)
plt.figure(figsize=(10, 8))
scatter = sns.scatterplot(data=df_flights, x='cost_usd', y='revenue_usd', hue='casm', palette='viridis', alpha=0.6)

plt.title('Cost vs. Revenue Colored by CASM', fontsize=16)
plt.xlabel('Cost (USD)', fontsize=12)
plt.ylabel('Revenue (USD)', fontsize=12)

# Add a y=x reference line
max_val = max(df_flights['cost_usd'].max(), df_flights['revenue_usd'].max())
plt.plot([0, max_val], [0, max_val], color='red', linestyle='--', label='Cost = Revenue')

plt.legend(title='CASM')
plt.tight_layout()
plt.show()
````
